```python
import pandas as pd
import numpy as np
```

# READING THE  DATA

```python
df_train=pd.read_csv('train_upvotes.csv')
```

```python
df_train
```

# CONVERTING CATEGORICAL DATA TO NUMERICAL

```python
encoded_df=pd.get_dummies(data=df_train,columns=['Tag'])
```

```python
encoded_df
```

# REMOVING ID COLUMN

```python
encoded_df.drop(['ID'],axis=1)
```

# SPLITTING THE DATA INTO TRAIN AND TEST 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(encoded_df.drop('Upvotes',axis=1), 
                                                    encoded_df['Upvotes'], test_size=0.20, 
                                                    random_state=101)
```

```python
import xgboost
classifier= xgboost.XGBRegressor()
classifier.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(classifier,X_test,y_test.ravel(),cv=10)
```

# THE FINAL SCORE

```python
score.mean()
```
